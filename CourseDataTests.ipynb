--Domain Integrity Queries
--(check that each column has sensible data types)

--Check to see that Catalog_ID and Title make sense given ProgramCode and ProgramNameb

```python
%%sql
SELECT Catalog_ID,Course_Title,ProgramCode,ProgramName
FROM PROGRAMS
    JOIN COURSE USING (PID)
    LIMIT 20;
```

--Check to see that see that Dawn W Massey comes up as teaching AC classes ~ note: this is also a relational integrity check as it assesses whether FID FK in the COURSE_OFFERING table corresponds to the FID PK in the FACULTY table

```python
%%sql
SELECT FACULTY.LName,FACULTY.FName,Title,Section,Semester,Year
FROM FACULTY
    JOIN COURSE_OFFERING USING(FID)
WHERE FACULTY.LName='Massey' and FACULTY.FName='Dawn W'
ORDER BY Year, Semester
;
```

--Note that the above is correct as Dawn W Massey also has taught Business Practicum for the last 4 semesters

--Relational Integrity Queries
--(check that the FKs are join compatible w/PKs; make sure mandatory relationships fulfill NOT NULL constraint)

--Make sure Title & Course_Title make sense as similar

```python
%%sql
SELECT CRN,Semester,Year,Title,COURSE.Catalog_ID,Course_Title
FROM COURSE_OFFERING
     JOIN COURSE USING(CID)
LIMIT 20;
```

--Make sure Catalog_ID & ProgramCode, ProgramName make sense as similar

```python
%%sql
SELECT Catalog_ID,ProgramCode,ProgramName
FROM COURSE
    JOIN PROGRAMS USING (PID)
LIMIT 20;
```

--Make sure Day,Start,End & Meetings coincide

```python
%%sql
SELECT CLASS_MEETINGS.CRN,Day,Start,End,Section,Meetings
FROM CLASS_MEETINGS
    JOIN COURSE_OFFERING USING (COID)
LIMIT 100;
```

--Want to see that Dawn W Massey comes up as teaching AC classes ~ note: this is also a “Domain Integrity Query” – i.e., assesses if the data “makes sense”

```python
%%sql
SELECT FACULTY.LName,FACULTY.FName,Title,Section,Semester,Year
FROM FACULTY
    JOIN COURSE_OFFERING USING(FID)
WHERE FACULTY.LName='Massey' and FACULTY.FName='Dawn W'
ORDER BY Year, Semester
;
```
